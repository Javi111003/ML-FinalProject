```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leer el archivo Excel
df = pd.read_excel("../data/muestra.xlsx")

# Definir el mapeo de categorías
category_map = {
    5: "Datos",
    2: "SMS",
    1: "Voz",
    15: "Recarga"
}

# Filtrar solo las categorías relevantes
df_filtered = df[df["SERVICE_CATEGORY"].isin(category_map.keys())]

# Reemplazar los códigos por nombres
df_filtered["Categoria"] = df_filtered["SERVICE_CATEGORY"].map(category_map)

# Contar la frecuencia de cada categoría
counts = df_filtered["Categoria"].value_counts()

# Crear gráfico de pastel
plt.figure(figsize=(6,6))
plt.pie(counts, labels=counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Proporción de categorías de servicio")
plt.savefig("results/grafico_pastel.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Contar clientes distintos por OBJ_ID
num_clientes = df_filtered["OBJ_ID"].nunique() 
print("Cantidad de clientes distintos:", num_clientes)
```
